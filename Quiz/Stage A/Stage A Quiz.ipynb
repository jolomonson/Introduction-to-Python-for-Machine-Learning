```python
import pandas as pd
df = pd.read_csv('dataset/FoodBalanceSheets.csv', encoding='latin-1', index_col=0)
df
```

```python
print('Question 1 - What is the total sum of AnimalFat produced in 2014 and 2017 respectively')
total_sum_of_animalFat = df.groupby(['Item'])['Y2014','Y2017'].sum()
print(total_sum_of_animalFat)
print('The total sum of Animal Fat is 209460.54 and 269617.53')
print('groupby() is basically a form of joining/grouping multiple columns using a central column provided as a parameter')
```

```python
print('Question 2 - What is the mean and standard deviation across the whole dataset for Y2015')
mean_of_Y2015 = df['Y2015'].mean()
std_of_Y2015 = df['Y2015'].std()
print('Mean of 2015 -',mean_of_Y2015)
print('Standard Deviation of 2015 -',std_of_Y2015)
```

```python
print('Question 3 - What is the total number and percentage of missing data in 2016')
missing_data = df['Y2016'].isnull().sum()
percentage = missing_data * 100 / len(df['Y2016'])
print('Total number of missing data',missing_data)
print('Percentage of missing data',percentage)
```

```python
print('Question 4 - What year had the highest correlation with Element Code')
highest_correlation = df.groupby(['Element Code']).count()
highest_correlation.corr()
print('The year with the highest correlation is 2017 with the value - 0.995815')
```

```python
print('Question 5 - What year has the highest sum of Import Quantity')
highest_sum_of_import_quantity = df.groupby(['Element'])['Y2014','Y2015','Y2016','Y2017','Y2018'].sum()
print(highest_sum_of_import_quantity)
print('The year with the highest sum of Import Quantity is Year 2017 with a sum of 294559.09')
```

```python
print('Question 6 - What is the total number of the sum of Production in 2014')
total_number_of_production = df.groupby(['Element'])['Y2014'].sum()
print(total_number_of_production)
print('The total number of the sum of Production in 2014 is - 1931287.75 as shown in the dataset below')
```

```python
print('Question 7 - Which of these elements had the highest sum in 2018')
highest_sum_in_2018 = df.groupby(['Element'])['Y2018'].sum()
print(highest_sum_in_2018)
print('The element with the highest sum is Domestic supply quantity with a value of -',highest_sum_in_2018.max())
```

```python
print('Question 8 -  Which of these elements had the 3rd Lowest sum in 2018')
third_lowest_sum_element_in_2018 = df.groupby(['Element'])['Y2018'].sum()
print(third_lowest_sum_element_in_2018)
print('The element with the lowest sum in 2018 is Tourist consumption with',third_lowest_sum_element_in_2018.min())
print('The element with the 3rd lowest sum in 2018 is Protein supply quantity (g/capita/day) with a value of 11833.56')
```

```python
print('Question 9 - What is the total Import Quantity in Algeria in 2018')
total_import_quantity = df.groupby(['Area','Element'])['Y2018'].sum()
total_import_quantity.head(10)
print('The total Import Quantity in Algeria in 2018 is 36238.29 as shown in the table below')
```

```python
print('Question 10 - What is the total number of unique countries in the dataset')
unique_countries = len(df['Area'].unique())
print('The total numbers of countries in the dataset is -',unique_countries)
```
